```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, FloatSlider

# Parameters
T = 2*np.pi      # period
N = 1000         # number of samples
t = np.linspace(0, T, N, endpoint=False)  # time vector
omega0 = 1       # base angular frequency
n_max = 12       # number of harmonics for Fourier series

# Function to compute Fourier coefficients for harmonics -n_max..n_max
def fourier_coefs(signal, t, omega0, n_max):
    dt = t[1] - t[0]
    n_range = np.arange(-n_max, n_max+1)
    coefs = []
    for n in n_range:
        c_n = np.sum(signal * np.exp(-1j*n*omega0*t)) * dt / (t[-1] + dt)
        coefs.append(c_n)
    return n_range, np.array(coefs)

def plot_signal(phi, n_signal):
    omega_signal = n_signal * omega0
    x = np.cos(omega_signal * t + phi)
    
    # Basis functions at the signal frequency
    cos_basis = np.cos(omega_signal * t)
    sin_basis = np.sin(omega_signal * t)
    
    # Fourier coefficients for -n_max .. n_max
    n_range, coefs = fourier_coefs(x, t, omega0, n_max)
    
    # Plot signal and basis
    plt.figure(figsize=(14,5))
    plt.subplot(1,2,1)
    plt.plot(t, x, label=fr'$cos({n_signal}\omega_0 t + \phi)$')
    plt.plot(t, cos_basis, '--', label=fr'$\cos({n_signal}\omega_0 t)$')
    plt.plot(t, sin_basis, '--', label=fr'$\sin({n_signal}\omega_0 t)$')
    plt.title(f'Signal and Basis Functions (phi={phi:.2f} rad, n={n_signal})')
    plt.xlabel('t')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)
    
    # Plot Fourier coefficients (amplitude) as stem
    plt.subplot(1,2,2)
    plt.stem(n_range, coefs.real, linefmt='b-', markerfmt='bo', basefmt=" ", label='Re(c_n)')
    plt.stem(n_range, coefs.imag, linefmt='r-', markerfmt='ro', basefmt=" ", label='Im(c_n)')
    plt.title('Fourier Coefficients (negative and positive n)')
    plt.xlabel('Harmonic n')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Interactive widget
interact(plot_signal, 
         phi=FloatSlider(value=0, min=0, max=2*np.pi, step=0.1, description='phi'),
         n_signal=IntSlider(value=3, min=1, max=12, step=1, description='n*omega0'));
```
