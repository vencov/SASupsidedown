# Fourier Series Approximation of a Sawtooth Signal

## Signal definition

We consider a periodic sawtooth signal with period $T_0$ and amplitude $A$:

$$
x(t) = \frac{2A}{T_0} t - A, \quad t \in [0, T_0)
$$

with periodic extension:

$$
x(t + T_0) = x(t)
$$


## Fourier series representation

The complex Fourier series is:

$$
x(t) = \sum_{n=-\infty}^{\infty} c_n e^{j n \omega_0 t}
$$

where the fundamental angular frequency is:

$$
\omega_0 = \frac{2\pi}{T_0}
$$


## Fourier coefficients (complex form)

The Fourier coefficients are:

$$
c_n =
\begin{cases}
0, & n = 0 \\
\frac{j A}{\pi n}, & n \neq 0
\end{cases}
$$



## Reconstruction using complex series

The signal is given by:

$$
x(t) = \sum_{n=-\infty}^{\infty} c_n e^{j n \omega_0 t}
$$

Since $c_0 = 0$, we can write:

$$
x(t) = \sum_{n \neq 0} \frac{j A}{\pi n} e^{j n \omega_0 t}
$$


Split the sum into positive and negative indices:

$$
x(t) = \sum_{n=1}^{\infty} \left( c_n e^{j n \omega_0 t} + c_{-n} e^{-j n \omega_0 t} \right)
$$

Using:

$$
c_n = \frac{jA}{\pi n}, \quad c_{-n} = \frac{jA}{\pi (-n)} = -\frac{jA}{\pi n}
$$

we get:

$$
x(t) = \sum_{n=1}^{\infty} \left( \frac{jA}{\pi n} e^{j n \omega_0 t} - \frac{jA}{\pi n} e^{-j n \omega_0 t} \right)
$$



Factor out constants:

$$
x(t) = \sum_{n=1}^{\infty} \frac{jA}{\pi n} \left( e^{j n \omega_0 t} - e^{-j n \omega_0 t} \right)
$$

Now use Euler’s identity:

$$
e^{j\theta} - e^{-j\theta} = 2j \sin(\theta)
$$

So:

$$
x(t) = \sum_{n=1}^{\infty} \frac{jA}{\pi n} \cdot (2j) \sin(n \omega_0 t)
$$

Since $j \cdot j = -1$:

$$
x(t) = \sum_{n=1}^{\infty} \frac{-2A}{\pi n} \sin(n \omega_0 t)
$$




```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

# Parameters
A = 1.0
T0 = 1.0
w0 = 2*np.pi / T0

# Time axis: 3 periods
t = np.linspace(-1.5*T0, 1.5*T0, 2000)

# Periodic sawtooth
def sawtooth(t):
    t_mod = t % T0   # now in [0, T0)
    return (2*A/T0) * t_mod - A

# Correct Fourier coefficients (fixed sign!)
def c_n(n):
    if n == 0:
        return 0
    else:
        return -A / (1j * np.pi * n)

# Fourier approximation
def fourier_approx(t, N):
    s = np.zeros_like(t, dtype=complex)
    for n in range(-N, N+1):
        s += c_n(n) * np.exp(1j * n * w0 * t)
    return s.real

# Interactive plot
def update(N=5):
    fig, axs = plt.subplots(1, 2, figsize=(13, 4))
    
    # --- Time domain ---
    x_true = sawtooth(t)
    x_approx = fourier_approx(t, N)
    
    axs[0].plot(t, x_true, label='Original', linewidth=2)
    axs[0].plot(t, x_approx, '--', label=f'Approx (N={N})')
    axs[0].set_title("Time domain (3 periods)")
    axs[0].legend()
    axs[0].grid(True)
    
    # --- Frequency domain ---
    n_full = np.arange(-20, 21)
    c_full = np.array([c_n(n) for n in n_full])
    
    # Mask for selected coefficients
    selected = (n_full >= -N) & (n_full <= N)
    
    # Plot ALL coefficients (faded)
    axs[1].stem(n_full, c_full.real, linefmt='C0-', markerfmt='C0o',
                basefmt=" ", label='Re{c_n} (all)')
    axs[1].stem(n_full, c_full.imag, linefmt='C1-', markerfmt='C1s',
                basefmt=" ", label='Im{c_n} (all)')
    
    # Reduce opacity for all stems
    for line in axs[1].lines:
        line.set_alpha(0.2)
    
    # Highlight SELECTED coefficients
    axs[1].stem(n_full[selected], c_full.real[selected],
                linefmt='C0-', markerfmt='C0o',
                basefmt=" ", label='Re{c_n} (used)')
    
    axs[1].stem(n_full[selected], c_full.imag[selected],
                linefmt='C1-', markerfmt='C1s',
                basefmt=" ", label='Im{c_n} (used)')
    
    axs[1].set_title("Fourier coefficients (highlighted selection)")
    axs[1].legend()
    axs[1].grid(True)
    
    plt.tight_layout()
    plt.show()

# Slider
interact(update, N=IntSlider(min=0, max=20, step=1, value=5));
```
