```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Parameters
T0 = 2*np.pi      # period
N = 1000          # samples
t = np.linspace(-T0/2, T0/2, N, endpoint=False)  # centered time vector
A = 1.0           # amplitude of rectangular pulse
n_max = 20        # number of harmonics
omega0 = 2*np.pi / T0  # fundamental frequency

# Original rectangular signal (centered at 0)
def rect_signal(t, shift=0):
    x = np.zeros_like(t)
    x[np.abs(t - shift) <= T0/4] = A
    return x

# Compute Fourier coefficients for n = -n_max .. n_max
def fourier_coefs(x, t, omega0, n_max):
    dt = t[1] - t[0]
    n_range = np.arange(-n_max, n_max+1)
    c_n = np.array([np.sum(x * np.exp(-1j*n*omega0*t)) * dt / T0 for n in n_range])
    return n_range, c_n

# Plotting function with interactive shift
def plot_rect_shift(shift):
    x = rect_signal(t, shift=shift)
    n_range, c_n = fourier_coefs(x, t, omega0, n_max)
    
    plt.figure(figsize=(12,5))
    
    # Left: rectangular signal
    plt.subplot(1,2,1)
    plt.plot(t, x, label=f'Rectangular pulse, shift={shift:.2f}')
    plt.title('Shifted Rectangular Signal')
    plt.xlabel('t')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.ylim(-0.1*A, 1.1*A)
    plt.legend()
    
    # Right: Fourier coefficients (real and imaginary parts)
    plt.subplot(1,2,2)
    plt.stem(n_range, c_n.real, linefmt='b-', markerfmt='bo', basefmt=" ", label='Re(c_n)')
    plt.stem(n_range, c_n.imag, linefmt='r-', markerfmt='ro', basefmt=" ", label='Im(c_n)')
    plt.title('Fourier Coefficients (Real & Imag)')
    plt.xlabel('Harmonic n')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Interactive widget for time shift
interact(plot_rect_shift,
         shift=FloatSlider(value=0, min=0, max=T0/4, step=0.05, description='Time Shift'));
```
