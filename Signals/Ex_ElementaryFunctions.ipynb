## Elementary Sinusoidal Functions

Below, we illustrate two fundamental sinusoidal functions:

1. $A \cdot \sin\left(\frac{2\pi}{T_0} t\right)$ — a sine wave with amplitude $A$ and period $T_0$.  
2. $A \cdot \cos\left(\frac{2\pi}{T_0} t\right)$ — a cosine wave with the same amplitude and period.  

In addition, a third signal is shown:

3. $A \cdot \sin\left(\frac{2\pi}{T_0} t + \phi\right)$ — a sine wave with an adjustable phase shift $\phi$.

The interactive widget below allows the phase $\phi$ to be varied continuously.  
As $\phi$ changes, the third signal smoothly transitions between the sine and cosine functions, illustrating the relationship between phase shifts and sinusoidal waveforms.


```python

# Parameters
A = 1.0
T0 = 0.01
fs = 5000

# Time axis
t = np.linspace(0, 2*T0, int(fs*2*T0), endpoint=False)

def plot_phase(phi=0):
    x_sin = A * np.sin(2 * np.pi / T0 * t)
    x_cos = A * np.cos(2 * np.pi / T0 * t)
    x_phi = A * np.sin(2 * np.pi / T0 * t + phi)
    
    plt.figure(figsize=(8, 3))
    plt.plot(t, x_sin, label=r'$A\sin\left(\frac{2\pi}{T_0} t\right)$')
    plt.plot(t, x_cos, '--', label=r'$A\cos\left(\frac{2\pi}{T_0} t\right)$')
    plt.plot(t, x_phi, ':', linewidth=2,
             label=rf'$A\sin\left(\frac{{2\pi}}{{T_0}} t + \phi\right),\ \phi={phi:.2f}$')
    
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.title('Effect of Phase on a Sinusoid')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

interact(
    plot_phase,
    phi=FloatSlider(
        min=-np.pi,
        max=np.pi,
        step=0.05,
        value=0,
        description='Phase φ [rad]'
    )
);
```

## Exponential Function with Time Shift

In this example, we consider the exponential function

$$
x(t) = A \, \exp\!\big(a\,(t + \tau)\big),
$$

where:
- $A$ controls the amplitude,
- $a$ determines the rate of exponential growth ($a > 0$) or decay ($a < 0$),
- $\tau$ is a time-shift parameter.

The interactive widget below allows all three parameters to be varied independently, illustrating how amplitude scaling, time shifting, and exponential growth or decay affect the signal shape.


```python

# Time axis
fs = 5000
t = np.linspace(-1, 1, int(fs*0.04), endpoint=False)

def plot_exponential(A=1.0, a=-50.0, tau=0.0):
    x = A * np.exp(a * (t + tau))
    
    plt.figure(figsize=(8, 3))
    plt.plot(t, x, linewidth=2,
             label=rf'$A e^{{a(t+\tau)}}$')
    
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.title(f'Exponential Signal  (A={A:.2f}, a={a:.1f}, τ={tau:.4f} s)')
    plt.grid(True)
    plt.xlim(-1,1)
    plt.ylim(-8,8)
    plt.legend()
    plt.tight_layout()
    plt.show()

interact(
    plot_exponential,
    A=FloatSlider(
        min=-2.0,
        max=2.0,
        step=0.1,
        value=1.0,
        description='Amplitude A'
    ),
    a=FloatSlider(
        min=-2.0,
        max=2.0,
        step=0.05,
        value=-2.0,
        description='Exponent a'
    ),
    tau=FloatSlider(
        min=-0.25,
        max=0.25,
        step=0.005,
        value=0.0,
        description='Shift τ [s]'
    )
);
```

## Time Shifting of the Unit Step Function

The unit step (Heaviside) function is defined as

$$
\eta(t) =
\begin{cases}
0, & t < 0 \\
1, & t \ge 0
\end{cases}
$$

A time-shifted step function is given by

$$
\eta(t + \tau),
$$

where $\tau$ is the time-shift parameter.

The interactive widget below allows $\tau$ to be varied continuously, illustrating how the step location moves along the time axis while the signal shape remains unchanged.


```python


# Time axis
fs = 5000
t = np.linspace(-0.02, 0.02, int(fs*0.04), endpoint=False)

# Unit step function
def unit_step(t):
    return np.where(t >= 0, 1.0, 0.0)

def plot_two_steps(tau1=0.0, tau2=0.0):
    x1 = unit_step(t + tau1)
    x2 = unit_step(t + tau2)
    
    plt.figure(figsize=(8, 3))
    plt.plot(t, x1, linewidth=2, label=rf'$u(t+\tau_1),\ \tau_1={tau1:.4f}$')
    plt.plot(t, x2, '--', linewidth=2, label=rf'$u(t+\tau_2),\ \tau_2={tau2:.4f}$')
    
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.title('Two Time-Shifted Unit Step Functions')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

interact(
    plot_two_steps,
    tau1=FloatSlider(
        min=-0.015,
        max=0.015,
        step=0.0005,
        value=0.0,
        description='Shift τ₁ [s]'
    ),
    tau2=FloatSlider(
        min=-0.015,
        max=0.015,
        step=0.0005,
        value=0.005,
        description='Shift τ₂ [s]'
    )
);
```

## Discrete-Time Exponential Signal

In discrete time, an exponential signal can be written as

$$
x[k] = A \, q^{k},
$$

where:
- $A$ is the amplitude,
- $q$ controls the behavior of the sequence:
  - $|q| < 1$ → exponential decay,
  - $|q| > 1$ → exponential growth,
  - $q < 0$ → alternating sign,
- $k$ is the discrete-time index.

The interactive widget below allows the parameters $A$ and $q$ to be varied, illustrating different types of discrete-time exponential sequences.


```python

# Discrete-time index
k = np.arange(-4, 10)

def plot_discrete_exponential(A=1.0, q=0.8):
    x = A * (q ** k)
    
    plt.figure(figsize=(8, 3))
    plt.stem(k, x)
    
    plt.xlabel('k')
    plt.ylabel('x[k]')
    plt.title(f'Discrete-Time Exponential  (A={A:.2f}, q={q:.2f})')
    plt.xlim(-4,10)
    #plt.ylim(-20,20)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

interact(
    plot_discrete_exponential,
    A=FloatSlider(
        min=-2.0,
        max=2.0,
        step=0.1,
        value=1.0,
        description='Amplitude A'
    ),
    q=FloatSlider(
        min=-1.5,
        max=1.5,
        step=0.05,
        value=0.8,
        description='Base q'
    )
);
```

```python
def geometric_visualization(r, N):
    k = np.arange(0, N+1)
    terms = r**k
    partial_sums = np.cumsum(terms)
    
    fig, axs = plt.subplots(1, 2, figsize=(12,5))
    
    # --- Left: Terms ---
    axs[0].stem(k, terms)
    axs[0].set_title("Geometric Terms: r^k")
    axs[0].set_xlabel("k")
    axs[0].set_ylabel("r^k")
    axs[0].grid(True)
    
    # --- Right: Partial Sums ---
    axs[1].plot(k, partial_sums, marker='o')
    axs[1].set_title("Partial Sum S_N")
    axs[1].set_xlabel("N")
    axs[1].set_ylabel("Sum")
    axs[1].grid(True)
    
    # Infinite sum (only if |r| < 1)
    if abs(r) < 1:
        infinite_sum = 1 / (1 - r)
        axs[1].axhline(infinite_sum, linestyle='--')
        axs[1].set_title(f"Partial Sum (Limit = {infinite_sum:.2f})")
    
    plt.tight_layout()
    plt.show()

interact(
    geometric_visualization,
    r=FloatSlider(value=0.5, min=-1.2, max=1.2, step=0.05),
    N=IntSlider(value=10, min=1, max=50, step=1)
);
```

## Scaling and shifting of the function

Below you can control three parameters which scale and shift the function


```python

# Original function
def f(x, a=2):
    return x * np.exp(-a * x) * unit_step(x)

# Parameters
a = 2
x = np.linspace(-10, 10, 1000)   # slightly extended for shifts

def plot_function(A, b, x0):
    y_original = f(x, a)
    y_transformed = A * f(b * (x + x0), a)

    plt.figure(figsize=(7,5))
    plt.plot(x, y_original, label="Original f(x)", linewidth=2)
    plt.plot(x, y_transformed, label="A·f(b(x + x0))", linewidth=2)

    plt.xlim(-10, 10)
    plt.ylim(-0.5, 0.5)   # fixed vertical range
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Original and Transformed Function")
    plt.legend()
    plt.grid(True)
    plt.show()

interact(
    plot_function,
    A=FloatSlider(value=1, min=-2, max=2, step=0.1),
    b=FloatSlider(value=1, min=-2, max=2, step=0.1),
    x0=FloatSlider(value=0, min=-6, max=6, step=0.1),
);
```

## Time Shifting of a Unit Pulse

In this example, we consider a unit pulse signal $u(t)$ defined over a finite time interval.
A time-shifted version of this signal is given by

$$
u(t + \tau),
$$

where $\tau$ is the time shift parameter.

The interactive widget below allows $\tau$ to be varied continuously.
Changing $\tau$ shifts the pulse along the time axis without altering its shape or amplitude, illustrating the effect of time shifting on a signal.


```python

# Time axis
fs = 5000
t = np.linspace(-0.02, 0.02, int(fs*0.04), endpoint=False)

# Unit pulse definition
def unit_pulse(t, width=0.004):
    return np.where(np.abs(t) <= width/2, 1.0, 0.0)

def plot_shifted_pulse(tau=0.0):
    x = unit_pulse(t + tau)
    
    plt.figure(figsize=(8, 3))
    plt.plot(t, x, linewidth=2, label=r'$u(t + \tau)$')
    
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.title(f'Time-Shifted Unit Pulse  (τ = {tau:.4f} s)')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

interact(
    plot_shifted_pulse,
    tau=FloatSlider(
        min=-0.01,
        max=0.01,
        step=0.0005,
        value=0.0,
        description='Shift τ [s]'
    )
);
```

## Euler’s Formula: Geometric and Trigonometric Interpretation

Euler’s formula relates the complex exponential to sine and cosine:

$$
e^{j\phi} = \cos(\phi) + j\sin(\phi).
$$

The left panel shows the complex exponential as a rotating vector on the unit circle.
The right panel shows the corresponding values of $\cos(\phi)$ and $\sin(\phi)$ as functions of the phase $\phi$.

The slider controls the phase $\phi$ and highlights the current position in both representations.


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider

def plot_euler(phi=0.0, a=1):
    phi_eff = a * phi
    z = np.exp(1j * phi_eff)
    
    fig, axes = plt.subplots(1, 2, figsize=(11, 4))
    
    # --- Left: Complex plane ---
    ax = axes[0]
    
    theta = np.linspace(0, 2*np.pi, 400)
    ax.plot(np.cos(theta), np.sin(theta),
            'k--', alpha=0.5, label='Unit circle')
    
    ax.arrow(0, 0, z.real, z.imag,
             head_width=0.05, length_includes_head=True,
             linewidth=2, color='C0',
             label=r'$e^{j a\phi}$')
    
    # Projections
    ax.plot([z.real, z.real], [0, z.imag],
            'C1:', linewidth=2, label='Imag part')
    ax.plot([0, z.real], [0, 0],
            'C0:', linewidth=2, label='Real part')
    
    ax.set_aspect('equal')
    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-1.2, 1.2)
    ax.axhline(0, color='gray', linewidth=0.5)
    ax.axvline(0, color='gray', linewidth=0.5)
    
    ax.set_xlabel('Real')
    ax.set_ylabel('Imaginary')
    ax.set_title('Complex Plane')
    
    # Legend outside the plot
    ax.legend(loc='center left', bbox_to_anchor=(1.05, 0.5))
    ax.grid(True)
    
    # --- Right: Cosine and Sine vs Phase ---
    ax2 = axes[1]
    
    phi_axis = np.linspace(-np.pi, np.pi, 400)
    ax2.plot(phi_axis, np.cos(phi_axis),
             label=r'$\cos(\phi)$')
    ax2.plot(phi_axis, np.sin(phi_axis),
             '--', label=r'$\sin(\phi)$')
    
    # Current phase marker
    ax2.axvline(phi_eff, color='k', linestyle=':',
                linewidth=1.5, label=r'Current $a\phi$')
    ax2.plot(phi_eff, np.cos(phi_eff), 'o')
    ax2.plot(phi_eff, np.sin(phi_eff), 'o')
    
    ax2.set_xlim(-np.pi, np.pi)
    ax2.set_ylim(-1.2, 1.2)
    ax2.set_xlabel('Phase [rad]')
    ax2.set_ylabel('Amplitude')
    ax2.set_title('Trigonometric Functions')
    ax2.legend()
    ax2.grid(True)
    
    fig.suptitle(
        rf'Euler’s Formula Visualization   ($a={a}$, $\phi={phi:.2f}$ rad)',
        fontsize=12
    )
    
    plt.tight_layout()
    plt.show()

interact(
    plot_euler,
    phi=FloatSlider(
        min=-np.pi,
        max=np.pi,
        step=0.05,
        value=0.0,
        description='Phase φ [rad]'
    ),
    a=IntSlider(
        min=-1,
        max=1,
        step=2,
        value=1,
        description='Direction a'
    )
)
```

## Discrete-Time Dirac (Kronecker) Delta – Theory


In discrete time, the Dirac delta becomes the **Kronecker delta**:

$$
\delta[n] =
\begin{cases}
1 & \text{if } n = 0 \\
0 & \text{if } n \ne 0
\end{cases}
$$

A shifted version is:

$$
\delta[n - \tau] =
\begin{cases}
1 & \text{if } n = \tau \\
0 & \text{otherwise}
\end{cases}
$$

Unlike the continuous Dirac delta, this is an ordinary sequence.

---


The most important property is the **sifting property**:

$$
\sum_{n=-\infty}^{\infty} f[n] \delta[n-\tau] = f[\tau]
$$

The delta "picks out" exactly one sample of the function.


$$
y[n] = f[n]\delta[n-\tau]
$$

This produces:

$$
y[n] =
\begin{cases}
f[\tau] & n=\tau \\
0 & \text{otherwise}
\end{cases}
$$

This is true discrete-time sampling.

Any discrete-time signal can be written as:

$$
f[n] = \sum_{k=-\infty}^{\infty} f[k] \delta[n-k]
$$

This means:

- The delta functions form a basis
- The coefficients are the signal values
- This is analogous to basis vectors in linear algebra


Continuous vs Discrete Comparison

Continuous-time:
$$
\int f(t)\delta(t-\tau) dt = f(\tau)
$$

Discrete-time:
$$
\sum f[n]\delta[n-\tau] = f[\tau]
$$

Integration becomes summation, but the idea is identical.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

# Discrete-time axis
n = np.arange(-20, 21)

# Example discrete-time function
def f(n):
    return np.exp(-0.1*n) * (n >= 0)   # causal decaying exponential

def plot_discrete(tau):
    delta = np.zeros_like(n)
    delta[n == tau] = 1  # Kronecker delta
    
    y_sum = f(n) * delta
    
    plt.figure(figsize=(8,5))
    
    # Original function
    plt.stem(n, f(n), linefmt='C0-', markerfmt='C0o', basefmt=" ", label="f[n]")
    
    # Dirac impulse
    plt.stem(n, delta, linefmt='C1-', markerfmt='C1o', basefmt=" ", label="δ[n - τ]")
    
    # Sum
    plt.stem(n, y_sum, linefmt='C2-', markerfmt='C2o', basefmt=" ", label="f[n] * δ[n - τ]")
    
    plt.xlim(-20, 20)
    plt.ylim(-0.2, 2)
    plt.xlabel("n")
    plt.ylabel("Amplitude")
    plt.title("Discrete-Time Function + Shifted Dirac Impulse")
    plt.legend()
    plt.grid(True)
    plt.show()

interact(
    plot_discrete,
    tau=IntSlider(value=0, min=-20, max=20, step=1)
);
```
