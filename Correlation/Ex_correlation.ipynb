## Autocorrelation of a Unit Pulse

Autocorrelation measures the similarity of a signal with a time-shifted version of itself.  
For a signal $n(t)$, the autocorrelation function is defined as:

$$
R(\tau) = \int_{-\infty}^{\infty} u(t + \tau)u^*(t)\, dt
$$

In this interactive example:

- The **left panel** shows a unit pulse $u(t)$ and its shifted version $u(t + \tau)$.
- The **right panel** shows the autocorrelation function $R(\tau)$.
- Moving the slider changes $\tau$, highlighting the point on the autocorrelation curve that corresponds to the current shift.

This illustrates how the overlap between the signal and its shifted copy determines the autocorrelation value.


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Time axis
fs = 5000
t = np.linspace(-0.02, 0.02, int(fs*0.04), endpoint=False)

# Unit pulse
def unit_pulse(t, width=0.004):
    return np.where(np.abs(t) <= width/2, 1.0, 0.0)

# Autocorrelation function
def autocorr(signal, t):
    dt = t[1] - t[0]
    r = np.correlate(signal, signal, mode='full') * dt
    tau = np.linspace(-t[-1]-dt, t[-1]+dt, len(r))
    return tau, r

# Original signal
x = unit_pulse(t, width=0.01)
tau_vals, Rxx = autocorr(x, t)

def plot_autocorr_interactive(tau_slider=0.0):
    # Shifted signal
    x_shifted = unit_pulse(t + tau_slider, width=0.01)
    
    fig, axes = plt.subplots(1, 2, figsize=(11, 4))
    
    # --- Left: signals ---
    ax = axes[0]
    ax.plot(t, x, linewidth=2, label='n(t)')
    ax.plot(t, x_shifted, '--', linewidth=2, label=f'n(t+τ), τ={tau_slider:.4f} s')
    ax.set_xlabel('Time [s]')
    ax.set_ylabel('Amplitude')
    ax.set_title('Unit Pulse and Shifted Version')
    ax.grid(True)
    ax.legend()
    
    # --- Right: autocorrelation ---
    ax2 = axes[1]
    ax2.plot(tau_vals, Rxx, label=r'$R(\tau)$')
    ax2.axvline(tau_slider, color='k', linestyle=':', linewidth=1.5, label='Current τ')
    # mark the current value on the curve
    # find closest index
    idx = (np.abs(tau_vals - tau_slider)).argmin()
    ax2.plot(tau_vals[idx], Rxx[idx], 'ro', label='Current R')
    
    ax2.set_xlabel('τ [s]')
    ax2.set_ylabel('Autocorrelation')
    ax2.set_title('Autocorrelation Function')
    ax2.grid(True)
    ax2.legend()
    
    plt.tight_layout()
    plt.show()

interact(
    plot_autocorr_interactive,
    tau_slider=FloatSlider(
        min=-0.02,
        max=0.02,
        step=0.0005,
        value=0.000,
        description='τ [s]'
    )
);
```

## Cross-Correlation of Two Unit Pulses

Cross-correlation measures the similarity between two signals as one is shifted relative to the other:

$$
R_{xy}(\tau) = \int_{-\infty}^{\infty} x(t)\, y(t + \tau)\, dt
$$

In this interactive example:

- **Left panel:** shows two unit pulses, $x(t)$ and $y(t)$, with independent amplitude, width, and time shift.  
- **Right panel:** shows the cross-correlation function $R_{xy}(\tau)$.  
- Sliders allow you to adjust **amplitude**, **width**, and **time shift** of each pulse.  
- The current $\tau$ is highlighted on the cross-correlation curve, showing how similarity changes as $y(t)$ moves relative to $x(t)$.


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Time axis
fs = 5000
t = np.linspace(-0.05, 0.05, int(fs*0.1), endpoint=False)

# Unit pulse function
def unit_pulse(t, width=0.004, A=1.0):
    return A * np.where(np.abs(t) <= width/2, 1.0, 0.0)

# Cross-correlation function
def crosscorr(x, y, t):
    dt = t[1] - t[0]
    r = np.correlate(x, y, mode='full') * dt
    tau = np.linspace(-t[-1]-dt, t[-1]+dt, len(r))
    return tau, r

def plot_crosscorr(A1=1.0, width1=0.004, tau1=0.0,
                   A2=1.0, width2=0.004, tau2=0.0,
                   tau_slider=0.0):
    # Signals
    x = unit_pulse(t - tau1, width=width1, A=A1)
    y = unit_pulse(t - tau2, width=width2, A=A2)
    
    tau_vals, Rxy = crosscorr(x, y, t)
    
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    
    # --- Left: signals ---
    ax = axes[0]
    ax.plot(t, x, linewidth=2, label='x(t)')
    ax.plot(t, y, '--', linewidth=2, label=f'y(t), τ2={tau2:.4f} s')
    ax.set_xlabel('Time [s]')
    ax.set_ylabel('Amplitude')
    ax.set_title('Two Unit Pulses')
    ax.grid(True)
    ax.legend()
    
    # --- Right: cross-correlation ---
    ax2 = axes[1]
    ax2.plot(tau_vals, Rxy, label=r'$R_{xy}(\tau)$')
    
    # current tau marker
    ax2.axvline(tau_slider, color='k', linestyle=':', linewidth=1.5, label='Current τ')
    idx = (np.abs(tau_vals - tau_slider)).argmin()
    ax2.plot(tau_vals[idx], Rxy[idx], 'ro', label='Current R')
    
    ax2.set_xlabel('τ [s]')
    ax2.set_ylabel('Cross-correlation')
    ax2.set_title('Cross-Correlation Function')
    ax2.grid(True)
    ax2.legend()
    
    plt.tight_layout()
    plt.show()

interact(
    plot_crosscorr,
    A1=FloatSlider(min=0.1, max=2.0, step=0.1, value=1.0, description='A1'),
    width1=FloatSlider(min=0.001, max=0.01, step=0.001, value=0.004, description='Width1'),
    tau1=FloatSlider(min=-0.02, max=0.02, step=0.0005, value=0.0, description='τ1'),
    A2=FloatSlider(min=0.1, max=2.0, step=0.1, value=1.0, description='A2'),
    width2=FloatSlider(min=0.001, max=0.01, step=0.001, value=0.004, description='Width2'),
    tau2=FloatSlider(min=-0.02, max=0.02, step=0.0005, value=0.005, description='τ2'),
    tau_slider=FloatSlider(min=-0.05, max=0.05, step=0.0005, value=0.0, description='τ')
)
```
